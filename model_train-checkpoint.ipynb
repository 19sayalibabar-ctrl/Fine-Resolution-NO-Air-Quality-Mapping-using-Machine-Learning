```python
# =============================================================
#  STEP 2 — TRAIN MODEL + PREDICT NO2 AT EVERY PUNE LOCATION
#  Inputs:
#    - data/output/training_data_500m.csv       (from step 1)
#    - data/output/satellite_500m_with_road.csv (from step 1)
#
#  Outputs:
#    - model_outputs/no2_model_500m.joblib
#    - model_outputs/metrics_500m.json
#    - model_outputs/plot_validation.png
#    - model_outputs/plot_feature_importance.png
#    - data/output/all_predictions_500m.csv     (NO2 at every pixel)
#    - map_outputs/map1_annual_500m.png
#    - map_outputs/map2_seasonal_500m.png
#    - map_outputs/map3_exceedance_500m.png
#    - map_outputs/map4_monthly_trend.png
#    - map_outputs/prediction_summary_500m.json
#
#  Run: python3 step2_train_and_predict.py
# =============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import matplotlib.patches as mpatches
import warnings, os, json, time
from datetime import datetime
from scipy import stats as scipy_stats
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from joblib import dump

warnings.filterwarnings("ignore")
os.makedirs("model_outputs",        exist_ok=True)
os.makedirs("map_outputs/daily",    exist_ok=True)
os.makedirs("map_outputs/seasonal", exist_ok=True)

# ── PATHS ──────────────────────────────────────────────────────
SAT_PATH    = "C:/Users/Admin/no2_update_scratch/data/output/satellite_500m_with_road.csv"
TRAIN_PATH  = "C:/Users/Admin/no2_update_scratch/data/output/training_data_500m.csv"


# ── MODEL CONFIG ───────────────────────────────────────────────
RF_PARAMS = {
    "n_estimators":      500,
    "max_depth":         10,
    "min_samples_leaf":  3,
    "random_state":      42,
    "n_jobs":            -1,
}
N_CV_FOLDS   = 5
SEASON_NAMES = {0:"Winter", 1:"Summer", 2:"Monsoon", 3:"Post-monsoon"}

# ── MAP CONFIG ─────────────────────────────────────────────────
LAT_MIN, LAT_MAX = 18.40, 18.80
LON_MIN, LON_MAX = 73.70, 74.15
WHO_LIMIT        = 25.0
NAAQS_LIMIT      = 80.0

STATIONS = {
    "Gavalinagar":   (18.63673, 73.82487),
    "Katraj Dairy":  (18.45445, 73.85416),
    "Park Wakad":    (18.59700, 73.76200),
    "SPPU":          (18.52900, 73.85100),
    "Thergaon":      (18.61100, 73.78400),
    "Bhumkar Nagar": (18.59800, 73.77300),
    "Hadapsar":      (18.50300, 73.93900),
    "Panchawati":    (18.53600, 73.82600),
    "Savta Mali":    (18.62800, 73.80600),
    "Nigdi":         (18.65400, 73.78800),
}

NO2_CMAP = mcolors.LinearSegmentedColormap.from_list(
    "no2_aqi",
    ["#00E400","#FFFF00","#FF7E00","#FF0000","#8F3F97","#7E0023"],
    N=256
)

TARGET = "NO2_ground_ugm3"

# Columns that are identifiers — never used as ML features
ID_COLS = ["date","station_name","match_dist_km","gnd_lat","gnd_lon",
           "NO2_ground_ugm3","NO2_ground_std","reading_count",
           "NO2_total_umol_m2"]


# =============================================================
#  STEP A — LOAD TRAINING DATA + BUILD FEATURE LIST
# =============================================================

def load_training_data():
    print("\n" + "="*60)
    print("  A — LOADING TRAINING DATA")
    print("="*60)

    df = pd.read_csv(TRAIN_PATH)
    print(f"  Rows:    {len(df)}")
    print(f"  Columns: {df.columns.tolist()}")

    # Build feature list: everything that is NOT an identifier
    feature_cols = [c for c in df.columns
                    if c not in ID_COLS and c != TARGET]

    # Verify no NaNs in features or target
    null_feats = df[feature_cols].isnull().sum()
    null_feats = null_feats[null_feats > 0]
    if len(null_feats) > 0:
        print(f"  ⚠️  Null values in features:\n{null_feats}")
        for col in null_feats.index:
            df[col] = df[col].fillna(df[col].median())
            print(f"     Filled {col} with median")

    X = df[feature_cols].copy()
    y = df[TARGET].copy()

    print(f"\n  Feature columns ({len(feature_cols)}):")
    for c in feature_cols:
        print(f"    {c}")
    print(f"\n  Target: {TARGET}")
    print(f"  Target mean:  {y.mean():.2f} µg/m³")
    print(f"  Target range: {y.min():.2f} – {y.max():.2f} µg/m³")

    return df, X, y, feature_cols


# =============================================================
#  STEP B — SAMPLE WEIGHTS (fix season imbalance)
# =============================================================

def compute_weights(df):
    season_counts = df["season"].value_counts()
    n_seasons     = df["season"].nunique()
    total         = len(df)
    weights = df["season"].map(
        lambda s: total / (n_seasons * season_counts[s])
    ).values

    print("\n  Sample weights by season:")
    for s, name in SEASON_NAMES.items():
        if s in season_counts:
            w = total / (n_seasons * season_counts[s])
            print(f"    {name:<14}  {season_counts[s]:>4} rows  weight={w:.2f}")
    return weights


# =============================================================
#  STEP C — 5-FOLD CROSS VALIDATION
# =============================================================

def run_cv(X, y, weights):
    print("\n" + "="*60)
    print(f"  C — {N_CV_FOLDS}-FOLD CROSS VALIDATION")
    print("="*60)

    kf           = KFold(n_splits=N_CV_FOLDS, shuffle=True, random_state=42)
    fold_metrics = []
    all_true, all_pred, all_fold_ids = [], [], []
    feat_imp_acc = np.zeros(X.shape[1])

    for fold, (tr_idx, val_idx) in enumerate(kf.split(X), 1):
        Xtr, Xval = X.iloc[tr_idx], X.iloc[val_idx]
        ytr, yval = y.iloc[tr_idx], y.iloc[val_idx]
        wtr        = weights[tr_idx]

        m = RandomForestRegressor(**RF_PARAMS)
        m.fit(Xtr, ytr, sample_weight=wtr)
        yp = np.maximum(m.predict(Xval), 0)

        r2   = r2_score(yval, yp)
        rmse = np.sqrt(mean_squared_error(yval, yp))
        mae  = mean_absolute_error(yval, yp)
        bias = float(np.mean(yp - yval))

        fold_metrics.append({"fold":fold,"R2":r2,"RMSE":rmse,"MAE":mae,"Bias":bias})
        all_true.extend(yval.tolist())
        all_pred.extend(yp.tolist())
        all_fold_ids.extend([fold]*len(yval))
        feat_imp_acc += m.feature_importances_

        print(f"  Fold {fold}:  R²={r2:.3f}  RMSE={rmse:.2f}  "
              f"MAE={mae:.2f}  Bias={bias:+.2f}")

    metrics_df = pd.DataFrame(fold_metrics)
    avg = metrics_df.mean(numeric_only=True)
    std = metrics_df.std(numeric_only=True)

    print(f"\n  ── Average ──────────────────────────────────")
    print(f"  R²   = {avg['R2']:.3f} ± {std['R2']:.3f}")
    print(f"  RMSE = {avg['RMSE']:.2f} ± {std['RMSE']:.2f} µg/m³")
    print(f"  MAE  = {avg['MAE']:.2f} ± {std['MAE']:.2f} µg/m³")
    print(f"  Bias = {avg['Bias']:+.2f} ± {std['Bias']:.2f} µg/m³")

    r2a = avg["R2"]
    grade = ("Excellent ✅" if r2a>=0.75 else "Good ✅" if r2a>=0.60
             else "Moderate ⚠️" if r2a>=0.45 else "Needs improvement 🔴")
    print(f"\n  Model quality: {grade}")

    return (np.array(all_true), np.array(all_pred),
            np.array(all_fold_ids), metrics_df, feat_imp_acc/N_CV_FOLDS)


# =============================================================
#  STEP D — TRAIN FINAL MODEL ON ALL DATA
# =============================================================

def train_final(X, y, weights, feature_cols):
    print("\n" + "="*60)
    print("  D — TRAINING FINAL MODEL (all data)")
    print("="*60)

    model = RandomForestRegressor(**RF_PARAMS)
    model.fit(X, y, sample_weight=weights)

    # Immediate fit verification
    test = model.predict(X.head(3))
    print(f"  Fit verified — sample preds: {np.round(test, 1)} ✅")

    dump(model, "model_outputs/no2_model_500m.joblib")
    print("  Saved → model_outputs/no2_model_500m.joblib")

    return model


# =============================================================
#  STEP E — SAVE VALIDATION PLOTS + METRICS
# =============================================================

def save_outputs(all_true, all_pred, all_fold_ids,
                 metrics_df, feat_imp, feature_cols):
    print("\n" + "="*60)
    print("  E — SAVING PLOTS + METRICS")
    print("="*60)

    plt.rcParams.update({"axes.facecolor":"#f9f9f9","figure.facecolor":"white",
                          "axes.grid":True,"grid.alpha":0.4})
    res = all_pred - all_true

    # ── Validation plot (3 panels) ─────────────────────────────
    fig, axes = plt.subplots(1, 3, figsize=(16, 5))
    fig.suptitle("Pune NO2 — 500m Model Validation",
                 fontsize=13, fontweight="bold")

    # Scatter
    axes[0].scatter(all_true, all_pred, c=all_fold_ids,
                    cmap="tab10", alpha=0.45, s=12, linewidths=0)
    lim = max(all_true.max(), all_pred.max()) * 1.05
    axes[0].plot([0,lim],[0,lim],"r--",lw=1.5,label="Perfect 1:1")
    r2   = r2_score(all_true, all_pred)
    rmse = np.sqrt(mean_squared_error(all_true, all_pred))
    axes[0].text(0.05,0.92,f"R²={r2:.3f}\nRMSE={rmse:.1f}µg/m³",
                 transform=axes[0].transAxes,fontsize=10,
                 bbox=dict(facecolor="white",alpha=0.8))
    axes[0].set_xlabel("Observed NO2 (µg/m³)")
    axes[0].set_ylabel("Predicted NO2 (µg/m³)")
    axes[0].set_title("A — Predicted vs Observed")
    axes[0].legend(fontsize=9)

    # Residuals
    axes[1].scatter(all_pred, res, alpha=0.35, s=10,
                    color="#2196F3", linewidths=0)
    axes[1].axhline(0, color="red", lw=1.5, ls="--")
    axes[1].axhline(res.mean(), color="orange", lw=1.2, ls=":",
                    label=f"Bias={res.mean():+.1f}")
    axes[1].fill_between([0,all_pred.max()],
                         res.mean()-res.std(), res.mean()+res.std(),
                         alpha=0.1, color="orange")
    axes[1].set_xlabel("Predicted NO2 (µg/m³)")
    axes[1].set_ylabel("Residual (µg/m³)")
    axes[1].set_title("B — Residuals vs Predicted")
    axes[1].legend(fontsize=9)

    # Histogram
    axes[2].hist(res, bins=40, color="#4CAF50", edgecolor="white", alpha=0.8)
    axes[2].axvline(0, color="red", lw=1.5, ls="--", label="Zero")
    axes[2].axvline(res.mean(), color="orange", lw=1.5, ls=":",
                    label=f"Mean={res.mean():+.1f}")
    axes[2].set_xlabel("Residual (µg/m³)")
    axes[2].set_title("C — Residual Distribution")
    axes[2].legend(fontsize=9)

    plt.tight_layout()
    plt.savefig("model_outputs/plot_validation.png",
                dpi=150, bbox_inches="tight")
    plt.close()
    print("  Saved → model_outputs/plot_validation.png")

    # ── Feature importance ─────────────────────────────────────
    fig2, ax = plt.subplots(figsize=(10, 6))
    feat_df  = pd.DataFrame({"feature": feature_cols,
                              "importance": feat_imp})\
                 .sort_values("importance", ascending=True)

    def get_color(f):
        if "NO2_trop" in f or "NO2_strat" in f: return "#E53935"
        if f in ["NDVI","elevation_m","population_density",
                 "land_surface_temp_c"]:             return "#1E88E5"
        if f in ["month","season","day_of_year",
                 "day_of_week","is_weekend"]:         return "#43A047"
        if "road" in f:                              return "#FF6F00"
        return "#FB8C00"

    colours = [get_color(f) for f in feat_df["feature"]]
    ax.barh(feat_df["feature"], feat_df["importance"],
            color=colours, alpha=0.85, edgecolor="white")
    for i,(v,_) in enumerate(zip(feat_df["importance"],feat_df["feature"])):
        ax.text(v+0.001, i, f"{v:.3f}", va="center", fontsize=8)
    ax.set_xlabel("Mean importance")
    ax.set_title("Feature Importance — 500m Fine Resolution Model",
                 fontweight="bold")

    handles = [plt.Rectangle((0,0),1,1,color=c,alpha=0.85)
               for c in ["#E53935","#1E88E5","#43A047","#FF6F00","#FB8C00"]]
    ax.legend(handles, ["Satellite NO2","Spatial","Temporal",
                         "Road density","Geometry"],
              fontsize=9, loc="lower right")
    plt.tight_layout()
    plt.savefig("model_outputs/plot_feature_importance.png",
                dpi=150, bbox_inches="tight")
    plt.close()
    print("  Saved → model_outputs/plot_feature_importance.png")

    # ── Save CV predictions CSV ────────────────────────────────
    pd.DataFrame({
        "fold":            all_fold_ids,
        "NO2_ground_true": np.round(all_true, 3),
        "NO2_pred":        np.round(all_pred, 3),
        "residual":        np.round(all_pred-all_true, 3),
    }).to_csv("model_outputs/cv_predictions.csv", index=False)
    print("  Saved → model_outputs/cv_predictions.csv")

    # ── Save metrics JSON ──────────────────────────────────────
    avg = metrics_df.mean(numeric_only=True)
    std = metrics_df.std(numeric_only=True)
    metrics_out = {
        "model": "RandomForest_500m",
        "n_features": len(feature_cols),
        "feature_cols": feature_cols,
        "target": TARGET,
        "cv_R2":   {"mean": round(float(avg["R2"]),   4), "std": round(float(std["R2"]),   4)},
        "cv_RMSE": {"mean": round(float(avg["RMSE"]), 4), "std": round(float(std["RMSE"]), 4)},
        "cv_MAE":  {"mean": round(float(avg["MAE"]),  4), "std": round(float(std["MAE"]),  4)},
        "cv_Bias": {"mean": round(float(avg["Bias"]), 4), "std": round(float(std["Bias"]), 4)},
    }
    with open("model_outputs/metrics_500m.json","w") as f:
        json.dump(metrics_out, f, indent=2)
    print("  Saved → model_outputs/metrics_500m.json")


# =============================================================
#  STEP F — PREDICT NO2 AT EVERY PIXEL IN PUNE
# =============================================================
# WHY chunks: 2.5M rows × ~17 features would need ~4GB RAM at
# once. Processing in 200k-row chunks keeps peak memory <1GB.

def predict_all_pixels(model, feature_cols):
    print("\n" + "="*60)
    print("  F — PREDICTING NO2 AT ALL 2.5M PIXELS")
    print("="*60)
    print("  Loading full satellite dataset...")

    CHUNK_SIZE = 200_000
    chunks_written = 0
    out_path       = "C:/Users/Admin/no2_update_scratch/data/output/all_predictions_500m.csv"
    header_written = False

    # Stream through the file in chunks to avoid loading 2.5M rows at once
    for chunk in pd.read_csv(SAT_PATH, chunksize=CHUNK_SIZE):
        # Fill any missing features with 0 (should not happen if step 1 ran)
        missing = [c for c in feature_cols if c not in chunk.columns]
        for col in missing:
            chunk[col] = 0.0

        X_chunk = chunk[feature_cols]
        preds   = np.maximum(model.predict(X_chunk), 0)
        chunk["NO2_pred_ugm3"] = np.round(preds, 3)

        out_cols = ["date","latitude","longitude",
                    "NO2_trop_umol_m2","NO2_pred_ugm3","season","month"]
        out_cols = [c for c in out_cols if c in chunk.columns]

        chunk[out_cols].to_csv(
            out_path,
            mode="a" if header_written else "w",
            header=not header_written,
            index=False
        )
        header_written = True
        chunks_written += 1

        total_rows = chunks_written * CHUNK_SIZE
        print(f"  Chunk {chunks_written}  ~{total_rows:,} rows processed  "
              f"mean={preds.mean():.1f} µg/m³")

    print(f"\n  All predictions saved → {out_path} ✅")

    # Load just enough to compute summary stats and maps
    print("  Loading predictions for map generation...")
    df_pred = pd.read_csv(out_path)
    print(f"  Total predicted rows: {len(df_pred):,}")
    print(f"  Unique pixels: {df_pred[['latitude','longitude']].drop_duplicates().shape[0]:,}")
    print(f"  NO2 mean: {df_pred['NO2_pred_ugm3'].mean():.1f} µg/m³")
    print(f"  NO2 range: {df_pred['NO2_pred_ugm3'].min():.1f} – "
          f"{df_pred['NO2_pred_ugm3'].max():.1f} µg/m³")
    return df_pred


# =============================================================
#  STEP G — GENERATE FINE RESOLUTION MAPS
# =============================================================

def style_ax(ax, title):
    ax.set_xlim(LON_MIN, LON_MAX)
    ax.set_ylim(LAT_MIN, LAT_MAX)
    ax.set_title(title, fontsize=10, fontweight="bold", pad=6)
    ax.set_xlabel("Longitude", fontsize=8)
    ax.set_ylabel("Latitude",  fontsize=8)
    ax.tick_params(labelsize=7)
    ax.grid(True, ls=":", alpha=0.2, lw=0.4)
    ax.text(LON_MIN+0.01, LAT_MAX-0.022,
            f"WHO {WHO_LIMIT} | NAAQS {NAAQS_LIMIT} µg/m³",
            fontsize=6, color="#555",
            bbox=dict(facecolor="white",alpha=0.7,edgecolor="none",
                      boxstyle="round,pad=0.2"))


def add_stations(ax, labels=True):
    for name,(lat,lon) in STATIONS.items():
        ax.scatter(lon,lat,c="black",s=30,marker="^",zorder=5,linewidths=0)
        if labels:
            ax.annotate(name,(lon,lat),xytext=(3,3),
                        textcoords="offset points",fontsize=5.5,color="#111",
                        bbox=dict(facecolor="white",alpha=0.7,edgecolor="none",
                                  boxstyle="round,pad=0.1"))


def generate_maps(df_pred):
    print("\n" + "="*60)
    print("  G — GENERATING FINE RESOLUTION MAPS")
    print("="*60)

    annual = df_pred.groupby(["latitude","longitude"],
                              as_index=False)["NO2_pred_ugm3"].mean()
    lats   = annual["latitude"].values
    lons   = annual["longitude"].values
    vals   = annual["NO2_pred_ugm3"].values
    vmax   = min(np.percentile(vals, 98), 200)

    # ── MAP 1: Annual mean — true 500m scatter ─────────────────
    print("  Map 1: Annual mean (500m scatter)...")
    fig, ax = plt.subplots(figsize=(11, 10))
    sc = ax.scatter(lons, lats, c=vals, cmap=NO2_CMAP,
                    s=0.4, vmin=0, vmax=vmax, alpha=0.9, linewidths=0)
    style_ax(ax,
        f"Pune — Annual mean surface NO2 (500m resolution)\n"
        f"{df_pred['date'].nunique()} satellite days  |  "
        f"{len(annual):,} unique pixels")
    add_stations(ax)
    cb = fig.colorbar(sc, ax=ax, fraction=0.03, pad=0.03)
    cb.set_label("Predicted NO2 (µg/m³)", fontsize=9)
    cb.ax.axhline(NAAQS_LIMIT, color="red",    lw=1.5, ls="--")
    cb.ax.axhline(WHO_LIMIT,   color="orange", lw=1.5, ls="--")
    mi = np.nanargmax(vals)
    ax.annotate(f"Max: {vals.max():.0f} µg/m³",
                (lons[mi],lats[mi]),xytext=(8,8),
                textcoords="offset points",fontsize=8,color="darkred",
                fontweight="bold",
                arrowprops=dict(arrowstyle="->",color="darkred",lw=0.8))
    plt.tight_layout()
    plt.savefig("map_outputs/map1_annual_500m.png",dpi=250,bbox_inches="tight")
    plt.close()
    print("  Saved → map_outputs/map1_annual_500m.png ✅")

    # ── MAP 2: Seasonal comparison ─────────────────────────────
    print("  Map 2: Seasonal comparison...")
    fig2, axes2 = plt.subplots(2,2,figsize=(14,12))
    fig2.suptitle("Pune NO2 — Seasonal patterns (500m resolution)\n"
                  "Predicted surface NO2 (µg/m³)",
                  fontsize=12,fontweight="bold",y=0.98)
    vmax_s = min(np.percentile(df_pred["NO2_pred_ugm3"],97),180)

    for idx,(s,sname) in enumerate(SEASON_NAMES.items()):
        ax2 = axes2.flat[idx]
        sub = df_pred[df_pred["season"]==s]
        if len(sub)==0:
            ax2.set_visible(False); continue
        smean = sub.groupby(["latitude","longitude"],
                             as_index=False)["NO2_pred_ugm3"].mean()
        n_days = sub["date"].nunique()
        sc2 = ax2.scatter(smean["longitude"],smean["latitude"],
                          c=smean["NO2_pred_ugm3"],cmap=NO2_CMAP,
                          s=0.3,vmin=0,vmax=vmax_s,linewidths=0,alpha=0.9)
        style_ax(ax2,
            f"{sname}  mean={smean['NO2_pred_ugm3'].mean():.1f} µg/m³  n={n_days}")
        add_stations(ax2,labels=(idx==0))
        fig2.colorbar(sc2,ax=ax2,fraction=0.03,pad=0.03)

    plt.tight_layout()
    plt.savefig("map_outputs/map2_seasonal_500m.png",dpi=200,bbox_inches="tight")
    plt.close()
    print("  Saved → map_outputs/map2_seasonal_500m.png ✅")

    # ── MAP 3: Exceedance ──────────────────────────────────────
    print("  Map 3: Exceedance...")
    exc = df_pred.groupby(["latitude","longitude"],as_index=False).apply(
        lambda g: pd.Series({
            "pct_who":   (g["NO2_pred_ugm3"]>WHO_LIMIT).mean()*100,
            "pct_naaqs": (g["NO2_pred_ugm3"]>NAAQS_LIMIT).mean()*100,
        })
    ).reset_index(drop=True)

    fig3,ax3s = plt.subplots(1,2,figsize=(15,7))
    fig3.suptitle("Pune NO2 — % days exceeding limits (500m resolution)",
                  fontsize=12,fontweight="bold")
    for ax3,(col,title,cmap3) in zip(ax3s,[
        ("pct_who",  f"WHO (25 µg/m³) % days exceeded","YlOrRd"),
        ("pct_naaqs",f"NAAQS (80 µg/m³) % days exceeded","OrRd"),
    ]):
        sc3 = ax3.scatter(exc["longitude"],exc["latitude"],
                          c=np.clip(exc[col],0,100),cmap=cmap3,
                          s=0.3,vmin=0,vmax=100,linewidths=0,alpha=0.9)
        style_ax(ax3,title)
        add_stations(ax3)
        cb3 = fig3.colorbar(sc3,ax=ax3,fraction=0.03,pad=0.03)
        cb3.set_label("% of days",fontsize=9)
        mi3 = exc[col].idxmax()
        ax3.annotate(f"Worst: {exc[col].max():.0f}%",
                     (exc.loc[mi3,"longitude"],exc.loc[mi3,"latitude"]),
                     xytext=(8,8),textcoords="offset points",fontsize=8,
                     color="darkred",fontweight="bold",
                     arrowprops=dict(arrowstyle="->",color="darkred",lw=0.8))
    plt.tight_layout()
    plt.savefig("map_outputs/map3_exceedance_500m.png",dpi=200,bbox_inches="tight")
    plt.close()
    print("  Saved → map_outputs/map3_exceedance_500m.png ✅")

    # ── MAP 4: Monthly trend ───────────────────────────────────
    print("  Map 4: Monthly trend...")
    df_pred["ym"] = pd.to_datetime(df_pred["date"]).dt.to_period("M").astype(str)
    monthly = df_pred.groupby("ym")["NO2_pred_ugm3"].agg(
        ["mean","std","min","max"]).reset_index().sort_values("ym")

    fig4, ax4 = plt.subplots(figsize=(13,5))
    x = range(len(monthly))
    ax4.fill_between(x,monthly["mean"]-monthly["std"],monthly["mean"]+monthly["std"],
                     alpha=0.18,color="#2196F3",label="±1 std")
    ax4.plot(x,monthly["mean"],"-o",color="#1565C0",lw=2.2,ms=5,label="Monthly mean")
    ax4.axhline(NAAQS_LIMIT,color="red",lw=1.5,ls="--",label=f"NAAQS {NAAQS_LIMIT}")
    ax4.axhline(WHO_LIMIT,color="orange",lw=1.5,ls="--",label=f"WHO {WHO_LIMIT}")
    ax4.set_xticks(list(x))
    ax4.set_xticklabels(monthly["ym"],rotation=45,ha="right",fontsize=9)
    ax4.set_ylabel("Predicted NO2 (µg/m³)")
    ax4.set_title("Pune — Monthly NO2 trend (500m resolution)",fontweight="bold")
    ax4.legend(fontsize=9)
    ax4.grid(True,ls=":",alpha=0.4)
    plt.tight_layout()
    plt.savefig("map_outputs/map4_monthly_trend.png",dpi=180,bbox_inches="tight")
    plt.close()
    print("  Saved → map_outputs/map4_monthly_trend.png ✅")

    # ── Summary JSON ───────────────────────────────────────────
    summary = {
        "resolution_m":      500,
        "unique_pixels":     int(len(annual)),
        "unique_dates":      int(df_pred["date"].nunique()),
        "mean_NO2_ugm3":     round(float(vals.mean()),2),
        "max_NO2_ugm3":      round(float(vals.max()),2),
        "pct_above_WHO":     round(float((df_pred["NO2_pred_ugm3"]>WHO_LIMIT).mean()*100),1),
        "pct_above_NAAQS":   round(float((df_pred["NO2_pred_ugm3"]>NAAQS_LIMIT).mean()*100),1),
        "seasonal_means":    {SEASON_NAMES[s]: round(float(
            df_pred[df_pred["season"]==s]["NO2_pred_ugm3"].mean()),1)
            for s in SEASON_NAMES if len(df_pred[df_pred["season"]==s])>0},
    }
    with open("map_outputs/prediction_summary_500m.json","w") as f:
        json.dump(summary,f,indent=2)

    print(f"\n  KEY RESULTS:")
    print(f"    Pixels mapped:     {len(annual):,}")
    print(f"    Mean Pune NO2:     {summary['mean_NO2_ugm3']} µg/m³")
    print(f"    Above WHO  (25):   {summary['pct_above_WHO']}%")
    print(f"    Above NAAQS (80):  {summary['pct_above_NAAQS']}%")
    print("    Seasonal means:")
    for s,v in summary["seasonal_means"].items():
        print(f"      {s:<16} {v} µg/m³")


# =============================================================
#  MAIN
# =============================================================

if __name__ == "__main__":
    t0 = time.time()
    print("\n" + "█"*60)
    print("  STEP 2 — TRAIN + PREDICT (500m Fine Resolution)")
    print("█"*60)

    df, X, y, feat_cols = load_training_data()
    weights = compute_weights(df)

    (all_true, all_pred, all_fold_ids,
     metrics_df, feat_imp) = run_cv(X, y, weights)

    model = train_final(X, y, weights, feat_cols)
    save_outputs(all_true, all_pred, all_fold_ids,
                 metrics_df, feat_imp, feat_cols)

    print("\n" + "█"*60)
    print("  TRAINING COMPLETE — predicting all pixels")
    print("█"*60)

    df_pred = predict_all_pixels(model, feat_cols)
    generate_maps(df_pred)

    elapsed = time.time() - t0
    print("\n" + "█"*60)
    print("  ALL DONE")
    print("█"*60)
    print(f"  Total time: {elapsed//60:.0f}m {elapsed%60:.0f}s")
    print("\n  model_outputs/")
    print("    no2_model_500m.joblib")
    print("    metrics_500m.json")
    print("    plot_validation.png")
    print("    plot_feature_importance.png")
    print("    cv_predictions.csv")
    print("\n  data/output/")
    print("    all_predictions_500m.csv  ← NO2 at every 500m pixel")
    print("\n  map_outputs/")
    print("    map1_annual_500m.png      ← fine resolution annual map")
    print("    map2_seasonal_500m.png")
    print("    map3_exceedance_500m.png")
    print("    map4_monthly_trend.png")
    print("    prediction_summary_500m.json")
    print("█"*60)
```
